# 공시지가 데이터

```python
init_price_li = []

for i in range(2007, 2024):
    init_price_li.append(pd.read_csv(f'../data/added_features/공시가/공시지가_{i}년.csv', encoding='cp949'))
```

```python
len(init_price_li)
```

```python
re = []
for i in range(17):
    tmp = pd.DataFrame(init_price_li[i].groupby('시군구명')['공시지가(원/㎡)'].mean()).reset_index()
    tmp.columns = ['시군구명', '공시지가 평균']
    re.append(tmp)
```

```python
re[0]
```

# 원본 데이터
- 년도에 따라 나눠서 진행

```python
df = pd.read_csv('../data/preprocessed/13_land_value_per_district.csv')
```

```python
year_split_df_list = []

for i in range(2007, 2024):
    year_split_df_list.append(df[df['계약년'] == i])
```

# 원본에 적용

```python
def mapping(init_price_df, origin_df):
    init_price_mapping = pd.Series({district: init_price for district, init_price in zip(init_price_df['시군구명'], init_price_df['공시지가 평균'])})

    origin_df['공시지가 평균'] = np.nan
    origin_df['공시지가 평균'] = origin_df['공시지가 평균'].fillna(origin_df['구']).map(init_price_mapping)

    return origin_df

for i in range(17):
    year_split_df_list[i] = mapping(re[i], year_split_df_list[i])
```

```python
year_split_df_list[0]
```

```python
df = pd.concat([splited_df for splited_df in year_split_df_list]).sort_index()
```
