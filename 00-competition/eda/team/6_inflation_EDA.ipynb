# 인플레이션 데이터 분석
- 기대 인플레이션율 사용

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc
```

```python
df = pd.read_csv('../data/added_features/기대인플레이션율(전국, 월)_10161008.csv')
df
```

```python
df.drop(['통계표', '단위', '변환', 'CSI코드'], axis=1, inplace=True)
df
```

```python
cols = df.columns.tolist()
cols = [int(col.replace('/', '')) for col in cols]
cols
```

```python
df.columns = cols
df
```

```python
inflation_df = pd.DataFrame(df.iloc[10])
inflation_df.columns = ['기대 인플레이션']
inflation_df
```

# 원본에 적용

```python
del df
gc.collect()
```

```python
df = pd.read_csv('../data/preprocessed/5_interest_rate.csv')
df
```

```python
inflation_df.reset_index(inplace=True)
inflation_df.columns = ['년월', '기대 인플레이션']

inflation_df.info()
```

```python
inflation_mapping = pd.Series({ym: inflation for ym, inflation in zip(inflation_df['년월'], inflation_df['기대 인플레이션'])})

df['기대 인플레이션'] = np.nan
df['기대 인플레이션'] = df['기대 인플레이션'].fillna(df['계약년월'].map(inflation_mapping))
```
