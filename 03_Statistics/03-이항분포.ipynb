```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

1. 이항분포를 따르는 랜덤 데이터의 히스토그램을 생성
2. 시각화

```python
## 10번 시도 중 성공할 확률이 70%이다.(10번 찼을 때 골이 들어갈 확률이 70% 즉, 각각의 시도마다 골을 넣을 확률이 70%이다.)
## 이를 1000번 반복.
## 각 실험에서는 주어진 10번의 시도 중에서 성공한 횟수가 기록되며 이런 실험을 1000번 반복하면, 1000개의 성공 횟수 데이터가 생성

sns.distplot(np.random.binomial(n=10, p=0.7, size=1000), hist=True, kde=False)  ## 각각의 10번 시도를 1000번 반복하여 이항분포를 생성
plt.show()
```
