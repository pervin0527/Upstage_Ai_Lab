```python
import pandas as pd

df = pd.read_csv('/Users/pervin0527/Upstage_Ai_Lab/03_Statistics/data/두_독립표본_t검정_실습.csv', sep=',', encoding='CP949')

df
```

```python
from scipy import stats
import numpy as np

class_A = df[df['회사'] == 1]
company1 = np.array(class_A['작동시간'])

class_B = df[df['회사'] == 2]
company2 = np.array(class_B['작동시간'])

## 두 표본간의 분산이 같은지 검사함. pvalue가 0.747이므로 귀무가설을 채택. 즉, 두 집단의 분산에 차이가 없다.(등분산)
stats.bartlett(company1, company2)
```

```python
## 독립표본 t검정(등분산이기 때문에 equal_var=True)
## 검정통계를 보면 pvalue가 0.11로, 유의수준보다 높은 확롤이다. 즉, 두 회사 제품에 차이가 없다는 귀무가설이 채택된다.
print(stats.ttest_ind(company1, company2, equal_var=True))
```
