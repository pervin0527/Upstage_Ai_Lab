```python
import pandas as pd
df = pd.read_csv('/Users/pervin0527/Upstage_Ai_Lab/03_Statistics/data/일원분산분석_실습.csv', encoding='CP949')
df
```

```python
import scipy as sp
import numpy as np
import pingouin as pg
df1 = df[['store', 'satisfaction']]
pd.options.display.float_format='{:3f}'.format
store = []
for i in range(1, 4):
    store.append(df1[df1['store']==i]['satisfaction'].values)
sp.stats.levene(store[0], store[1], store[2])
```

```python
print(pg.anova(dv='satisfaction', between='store', data=df1))
```

```python
print(pg.welch_anova(dv='satisfaction', between='store', data=df1))
```
