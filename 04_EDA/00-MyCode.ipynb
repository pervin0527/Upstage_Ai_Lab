```python
left = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': [1, 2, 3, 4]})
print(left.head())

right = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value': [5, 6, 7, 8]})
print(right.head())
```

```python
# 'key' 열을 기준으로 내부 병합 (inner join)
merged_inner = pd.merge(left, right, on='key', how='inner')
print(merged_inner)
```

```python
# 'key' 열을 기준으로 외부 병합 (outer join)
merged_outer = pd.merge(left, right, on='key', how='outer')
print(merged_outer)
```

```python
merged_left = pd.merge(left, right, on='key', how='left')
print(merged_left)
```

```python
merged_right = pd.merge(left, right, on='key', how='right')
print(merged_right)
```

```python
left = pd.DataFrame({'left_key': ['A', 'B', 'C', 'D'],
                     'value': [1, 2, 3, 4]})
right = pd.DataFrame({'right_key': ['B', 'D', 'E', 'F'],
                      'value': [5, 6, 7, 8]})

merged = pd.merge(left, right, left_on='left_key', right_on='right_key', how='outer')
print(merged)
```

```python
left = pd.DataFrame({'value': [1, 2, 3, 4]}, index=['A', 'B', 'C', 'D'])
right = pd.DataFrame({'value': [5, 6, 7, 8]}, index=['B', 'D', 'E', 'F'])

merged = pd.merge(left, right, left_index=True, right_index=True, how='outer')
print(merged)
```

```python
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9],
                    'B': [10, 11, 12]})

concatenated = pd.concat([df1, df2], axis=0)
print(concatenated)
```

```python
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [7, 8, 9],
                    'B': [10, 11, 12]})
df3 = pd.DataFrame({'A': [13, 14, 15],
                    'C': [16, 17, 18]})

concatenated = pd.concat([df1, df3], join='inner')
print(concatenated)
```

```python
concatenated = pd.concat([df1, df3], join='outer')
print(concatenated)
```
