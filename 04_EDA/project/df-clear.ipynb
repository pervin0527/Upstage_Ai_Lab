# 4. 포지션 정제 및 세분화

```python
player_stats_df = [pl_player_stats_df, l1_player_stats_df, gb_player_stats_df, sa_player_stats_df, ll_player_stats_df]
```

```python
unique_positions_set = set()
for df in player_stats_df:
    unique_positions = df['position'].unique()
    unique_positions_set.update(unique_positions)

position_words = set()
for position in unique_positions_set:
    parts = position.split(',')
    for part in parts:
        cleaned_part = part.strip()
        position_words.add(cleaned_part)

print(list(position_words))
```

```python
def map_positions(pos):
    positions = [p.strip() for p in pos.split(',')]
    mapped_positions = [POS_FILTER.get(p, p) for p in positions]
    return ', '.join(mapped_positions)

df['position'] = df['position'].apply(map_positions)
print(player_stats_df[0]['position'].head(20))
```

```python
pl_player_stats_df['position'] = pl_player_stats_df['position'].apply(map_positions)
l1_player_stats_df['position'] = l1_player_stats_df['position'].apply(map_positions)
gb_player_stats_df['position'] = gb_player_stats_df['position'].apply(map_positions)
sa_player_stats_df['position'] = sa_player_stats_df['position'].apply(map_positions)
ll_player_stats_df['position'] = ll_player_stats_df['position'].apply(map_positions)
```

```python
unique_positions_set = set()
for df in player_stats_df:
    unique_positions = df['position'].unique()
    unique_positions_set.update(unique_positions)

position_words = set()
for position in unique_positions_set:
    parts = position.split(',')
    for part in parts:
        cleaned_part = part.strip()
        position_words.add(cleaned_part)

print(list(position_words))
```

```python
def split_positions(df):
    position_split = df['position'].str.split(',', expand=True)
    position_split.columns = [f'position_{i+1}' for i in range(position_split.shape[1])]
    result_df = pd.concat([df, position_split], axis=1)
    
    return result_df

pl_player_stats_df = split_positions(pl_player_stats_df)
l1_player_stats_df = split_positions(l1_player_stats_df)
gb_player_stats_df = split_positions(gb_player_stats_df)
sa_player_stats_df = split_positions(sa_player_stats_df)
ll_player_stats_df = split_positions(ll_player_stats_df)
```

# 5. 선수 데이터프레임의 팀명을 팀 데이터프레임의 이름으로 변경

```python
def team_name_mapper(player_df, team_df):
    player_teams_unique = sorted(player_df['team_name'].unique())
    team_teams_unique = sorted(team_df['team_name'].unique())

    team_name_dict = {}
    for pt, tt in zip(player_teams_unique, team_teams_unique):
        team_name_dict[pt] = tt
        # print(f"{pt} : {tt}")

    return team_name_dict


mapping_dict = team_name_mapper(pl_player_stats_df, pl_team_stats_df)
for key, value in mapping_dict.items():
    print(key, value)
pl_player_stats_df['team_name'] = pl_player_stats_df['team_name'].map(mapping_dict)
```
