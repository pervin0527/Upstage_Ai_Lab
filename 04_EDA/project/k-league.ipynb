```python
import os
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
import requests
from bs4 import BeautifulSoup

frame_url = 'https://fbref.com/en/comps/55/2023/2023-K-League-1-Stats'
browser = webdriver.Chrome()
browser.get(frame_url)
```

```python
container = browser.find_element(By.ID, 'results2023551_overall')
headers = container.find_element(By.TAG_NAME, 'thead').find_elements(By.TAG_NAME, 'th')
headers = [header.text for header in headers]

headers.remove('Attendance')
headers.append('URL')
```

```python
body = container.find_element(By.TAG_NAME, 'tbody')
rows = body.find_elements(By.TAG_NAME, 'tr')
del rows[6] ## blank row 제거.

team_stat_dataset = []
for row in rows:
    anchor = row.find_element(By.CLASS_NAME, 'left').find_element(By.TAG_NAME, 'a')
    team_name = anchor.text
    team_url = anchor.get_attribute('href')

    team_stat_data = row.find_elements(By.CLASS_NAME, 'right')
    team_stat_data = [data.text for data in team_stat_data]
    team_stat_data.insert(0, team_name)
    team_stat_data.append(team_url)
    team_stat_data.remove('')

    team_stat_dataset.append(team_stat_data)

df1 = pd.DataFrame(team_stat_dataset, columns=headers)
df1.head()
```

```python
standard_stat_table = browser.find_element(By.ID, 'stats_squads_standard_for')
standard_stat_header = standard_stat_table.find_element(By.TAG_NAME, 'thead')

stat_header = standard_stat_header.find_elements(By.TAG_NAME, 'tr')[1]
stat_headers = stat_header.find_elements(By.TAG_NAME, 'th')
for header in stat_headers:
    print(header.text)
```

```python
stat_body = standard_stat_table.find_element(By.TAG_NAME, 'tbody')
stat_body_rows = stat_body.find_elements(By.TAG_NAME, 'tr')

team_stat_dataset = []
for row in stat_body_rows:
    anchor = row.find_element(By.CLASS_NAME, 'left').find_element(By.TAG_NAME, 'a')
    team_name = anchor.text
    team_url = anchor.get_attribute('href')

    right_data = row.find_elements(By.CLASS_NAME, 'right')
    center_data = row.find_elements(By.CLASS_NAME, 'center')
    team_stat_data = right_data + center_data

    team_stat_data = [x.text for x in team_stat_data]
    team_stat_data.insert(0, team_name)
    team_stat_data.append(team_url)

    team_stat_dataset.append(team_stat_data)

stat_df = pd.DataFrame(team_stat_dataset, stat_headers)
stat_df.head()
```
