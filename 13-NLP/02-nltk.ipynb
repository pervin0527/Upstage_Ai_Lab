### 1-1. ì–´íœ˜ë¶„ì„

> ë¬¸ì¥ë‚´ì— ìˆëŠ” ì—¬ëŸ¬ ë‹¨ì–´ë“¤ì„ í˜•íƒœì†Œ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ê±°ë‚˜ ì–¸ì–´í•™ì ìœ¼ë¡œ ì–´ë–¤ í’ˆì‚¬ì— í•´ë‹¹í•˜ëŠ”ì§€ë¥¼ ë¶„ì„í•©ë‹ˆë‹¤. ë˜í•œ konlpyë¥¼ ì´ìš©í•´ í•œêµ­ì–´ì—ì„œ ì‚¬ìš©ë˜ëŠ” ì—¬ëŸ¬ í˜•íƒœì†Œ ë¶„ì„ê¸°ë¥¼ ì•Œì•„ë³´ê³  ì°¨ì´ì ì„ í™•ì¸í•©ë‹ˆë‹¤



**pos_tag**
+ nltkì˜ pos_tagëŠ” ë¶„ë¦¬ëœ ë‹¨ì–´ë“¤ì— ëŒ€í•´ì„œ í˜•íƒœì†Œ ë¶„ì„ í›„ ì•Œë§ëŠ” í’ˆì‚¬ë¥¼ íƒœê¹…í•´ì„œ ë³´ì—¬ì¤ë‹ˆë‹¤
+ ë”°ë¼ì„œ inputìœ¼ë¡œëŠ” ë¶„ë¦¬ëœ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ê°€ ì…ë ¥ë˜ì–´ì•¼í•˜ë©° outputìœ¼ë¡œëŠ” tupleí˜•íƒœì˜ `(í˜•íƒœì†Œ, í’ˆì‚¬)` í˜•íƒœë¡œ ë°˜í™˜ë©ë‹ˆë‹¤

````python
sentence="My name is jack and i am 22 years old and my major is natural language processing"

tokens = nltk.word_tokenize(sentence) # í† í¬ë‚˜ì´ì €ë¡œ ë¶„ë¦¬
tagged = nltk.pos_tag(tokens) # pos tagging ì‹¤í–‰

for i,j in tagged:
  print(f'{i} - {j}')
````

ì—¬ê¸°ì„œëŠ” pos_tag í•¨ìˆ˜ë¥¼ ì´ìš©í•´ ì›í•˜ëŠ” í’ˆì‚¬ì˜ í˜•íƒœì†Œë§Œ ì¶”ì¶œí•©ë‹ˆë‹¤

````python
allnoun = []
for word, pos in tagged:
  if pos in ['NN', 'NNP']: # í’ˆì‚¬íƒœê·¸ ì¤‘ NN, NNPë§Œ ì¶”ì¶œ
    allnoun.append(word)

print("í˜•íƒœì†Œ ë¶„ì„ ê²°ê³¼: ",tagged)
print("ëª…ì‚¬ë§Œ ì¶”ì¶œ: ", allnoun)

allverb = []
for word, pos in tagged:
  if pos in ['VBZ','VBD']: # í’ˆì‚¬íƒœê·¸ ì¤‘ VBZ, VBDë§Œ ì¶”ì¶œ
    allverb.append(word)

print("ë™ì‚¬ë§Œ ì¶”ì¶œ",allverb)
````

### 1-2. êµ¬ë¬¸ ë¶„ì„

> êµ¬ë¬¸ ë¶„ì„ì€ ëª…ì‚¬êµ¬, ë™ì‚¬êµ¬ ì²˜ëŸ¼ êµ¬ë¬¸êµ¬ì¡°ë¥¼ ë¶„ì„í•˜ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ë¬¸ì¥ì„ ì´ë£¨ê³  ìˆëŠ” êµ¬, ë‹¨ì–´ë“¤ì´ ë¬¸ì¥ ë‚´ì—ì„œ ì–´ë–¤ ìˆ˜ì‹, ì˜ì¡´ ê´€ê³„ë¥¼ ê°–ëŠ”ì§€ ë‚˜íƒ€ë‚´ëŠ” ì¼ì¢…ì˜ Treeë¥¼ ë§Œë“¤ì–´ì£¼ëŠ” ê±°ë¼ê³  ìƒê°í•˜ë©´ ì´í•´í•˜ê¸° ì‰½ìŠµë‹ˆë‹¤


````python
import konlpy
import nltk
````

êµ¬ë¬¸ ë¶„ì„ì„ ìœ„í•´ ìš°ì„  í˜•íƒœì†Œ ë¶„ì„ì„ ë¨¼ì € ìˆ˜í–‰í•©ë‹ˆë‹¤. í˜•íƒœì†Œ ë¶„ì„í•´ ë‹¨ì–´ë“¤ì˜ í’ˆì‚¬ë¥¼ ëª¨ë‘ íƒœê¹…í•©ë‹ˆë‹¤

````python
sentence = 'ë§Œ 6ì„¸ ì´í•˜ì˜ ì´ˆë“±í•™êµ ì·¨í•™ ì „ ìë…€ë¥¼ ì–‘ìœ¡í•˜ê¸° ìœ„í•´ì„œëŠ”'
words = konlpy.tag.Twitter().pos(sentence) # í’ˆì‚¬íƒœê¹…
words
````

````python
# ëª…ì‚¬, ë™ì‚¬, í˜•ìš©ì‚¬ êµ¬ì˜ ê·œì¹™ì„ ì •ì˜í•©ë‹ˆë‹¤
grammar = """
NP: {<N.*>*<Suffix>?}   # Noun phrase
VP: {<V.*>*}            # Verb phrase
AP: {<A.*>*}            # Adjective phrase
"""
parser = nltk.RegexpParser(grammar) # ì„ ì–¸ëœ grammer ruleì„ parserì—ê²Œ ì…ë ¥
chunks = parser.parse(words) # grammer ruleì— ë”°ë¼ ë‹¨ì–´ë¥¼ íŒŒì‹±
print("# Print whole tree")
print(chunks.pprint())

# ëª…ì‚¬êµ¬ì— í•´ë‹¹í•˜ëŠ” subtreeë§Œ ì¶œë ¥í•©ë‹ˆë‹¤
print("\n# Print noun phrases only")
for subtree in chunks.subtrees():
    if subtree.label()=='NP': # ì„œë¸ŒíŠ¸ë¦¬ì˜ í’ˆì‚¬ê°€ NPì¸ ê²½ìš°
        print(' '.join((e[0] for e in list(subtree))))
        print(subtree.pprint())
````

êµ¬ë¬¸ ë¶„ì„ì€ NLTK ë¿ë§Œì•„ë‹ˆë¼ Spacyë¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬ì—ì„œë„ ì†ì‰½ê²Œ ìˆ˜í–‰ì´ ê°€ëŠ¥í•©ë‹ˆë‹¤. ë¶€ê°€ì ìœ¼ë¡œ SpacyëŠ” ì˜ì–´ í˜•íƒœì†Œ ë¶„ì„ê¸°ë¡œì¨ NLTKë§Œí¼ ìì£¼ ì‚¬ìš©ë˜ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬ë¡œ êµ¬ë¬¸ë¶„ì„ì— ìˆì–´ì„œ NLTKë³´ë‹¤ ì§ê´€ì ì¸ ì¥ì ì´ìˆê¸° ë•Œë¬¸ì—, ì´ë²ˆ ì‹¤ìŠµ ì—¬ê¸° ë¶€ë¶„ì—ì„œë§Œ ì ê¹ ë‹¤ë£¨ì–´ ë³´ê² ìŠµë‹ˆë‹¤

## 2. NLTKë¥¼ í™œìš©í•œ í‚¤ì›Œë“œ ë¶„ì„

```
ğŸ’¡ ëª©ì°¨ ê°œìš” : NLTK ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì´ìš©í•´ corpusì—ì„œ ë¹ˆë„ê¸°ë°˜ í‚¤ì›Œë“œ ë¶„ì„ì„ ì‹¤í–‰í•´ ë´…ë‹ˆë‹¤
```

- ë¬¸ìì—´ corpusë‚´ì—ì„œ í˜•íƒœì†Œ ë¶„ì„ì„ í†µí•´ ëª…ì‚¬, í˜•ìš©ì‚¬ë§Œì„ ì¶”ì¶œí•´ ë¹ˆë„ìˆ˜ ê·¸ë˜í”„ì™€ ì›Œë“œ í´ë¼ìš°ë“œë¥¼ ìƒì„±í•©ë‹ˆë‹¤

````python
import pandas as pd
from tqdm import tqdm
from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud

import nltk
from nltk.corpus import stopwords
from nltk.tag import pos_tag
from nltk import word_tokenize
nltk.download('book', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('stopwords')

nltk.corpus.gutenberg.fileids() # nltk.corpus.gutenberg ëŠ” ì—¬ëŸ¬ í…ìŠ¤íŠ¸ ì½”í¼ìŠ¤ë“¤ì„ ì œê³µ
````

corpusë¥¼ word_tokenize í•¨ìˆ˜ë¥¼ í†µí•´ í˜•íƒœì†Œ ë¶„ì„ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤

extract_by_tagëŠ” word_listì™€ tag_nameì„ ì…ë ¥ë°›ì•„ word_listì— ëŒ€í•´ ì›í•˜ëŠ” í’ˆì‚¬ íƒœê·¸ë¥¼ tag_nameì— ì…ë ¥í•˜ë©´ í•´ë‹¹ í’ˆì‚¬ì— í•´ë‹¹í•˜ëŠ” ë‹¨ì–´ë“¤ë§Œ ì¶”ì¶œì´ ê°€ëŠ¥í•œ í•¨ìˆ˜ì…ë‹ˆë‹¤. sortingì€ Counter ëª¨ë“ˆì„ ì´ìš©í•´ ë‹¨ì–´ë¥¼ ë¹ˆë„ìˆ˜ê¸°ë°˜ìœ¼ë¡œ ì •ë ¬í•´ì¤ë‹ˆë‹¤

````python
# ìœ„ì—ì„œ ì¶”ì¶œëœ morphsë¥¼ í†µí•´ ì›í•˜ëŠ” í’ˆì‚¬ì— í•´ë‹¹í•˜ëŠ” ë‹¨ì–´ë“¤ë§Œ ì¶”ì¶œí•´ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜
def extract_by_tag(word_list, tag_name):
  pos_list = pos_tag(word_list, tagset='universal') # ì—¬ê¸°ì„œëŠ” universal tagë¥¼ ì´ìš©
  extracted_list = []
  for word_tag in tqdm(pos_list):
    if tag_name in word_tag[1]: # 0ë²ˆì€ í’ˆì‚¬ íƒœê·¸. 1ë²ˆì€ í…ìŠ¤íŠ¸ë¡œ ì—¬ê¸°ì„œëŠ” í…ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ
      extracted_list.append(word_tag[0])
  return extracted_list

# ì¶”ì¶œëœ ë¦¬ìŠ¤íŠ¸ë¥¼ ì¶œí˜„ë‹¨ì–´ ë¹ˆë„ìˆ˜ë¥¼ ë§¤í•‘í•˜ì—¬ ì •ë ¬
def sorting(sentence_list):
  count = Counter(sentence_list)
  words = (dict(count.most_common()))
  return words
````

````python
nouns = extract_by_tag(word_list,'NOUN') # ëª…ì‚¬ë§Œ ì¶”ì¶œ
print(f'ëª…ì‚¬ ì˜ˆì‹œ:{nouns[5:10]}')

adjs = extract_by_tag(word_list,'ADJ') # í˜•ìš©ì‚¬ë§Œ ì¶”ì¶œ
print(f'\ní˜•ìš©ì‚¬ ì˜ˆì‹œ:{adjs[:5]}')
````

nltkì—ì„œ ì •ì˜ëœ ë¶ˆìš©ì–´ë“¤ì„ ëª…ì‚¬, í˜•ìš©ì‚¬ ë¦¬ìŠ¤íŠ¸ì—ì„œ ì œì™¸í•˜ëŠ” ì •ì œì‘ì—…ì„ ê±°ì¹©ë‹ˆë‹¤.

ê·¸ í›„ sortingì„ í†µí•´ ë†’ì€ ë¹ˆë„ìˆ˜ë¥¼ ê°–ëŠ” ë‹¨ì–´ ìˆœìœ¼ë¡œ ì •ë ¬í•´ì¤ë‹ˆë‹¤

````python
# ì œê±°í•  ë¶ˆìš©ì–´ë¥¼ ì •ì˜
stop = stopwords.words('english')
stop.extend(['mr.','mrs.']) # ì›í•˜ëŠ” ë¶ˆìš©ì–´ ì¶”ê°€

# ì§€ì •í•œ ë¶ˆìš©ì–´ë¥¼ ê°ê° ë¦¬ìŠ¤íŠ¸ì—ì„œ ì œê±°í•¨
nouns = [token for token in nouns if not token in stop]
adjs = [token for token in adjs if not token in stop]

# ì¶”ì¶œëœ ë¦¬ìŠ¤íŠ¸ë¥¼ ì¶œí˜„ë‹¨ì–´ ë¹ˆë„ìˆ˜ë¥¼ ë§¤í•‘í•˜ì—¬ ì •ë ¬
nouns_count = sorting(nouns)
adj_count = sorting(adjs)
````

ë§ˆì§€ë§‰ìœ¼ë¡œ Wordcloudì™€ ë¹ˆë„ê¸°ë°˜ í‚¤ì›Œë“œ ê·¸ë˜í”„ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. WordCloud í•¨ìˆ˜ì— generate_from_frequenciesë¥¼ ì‚¬ìš©í•´ Counter í´ë˜ìŠ¤ë¡œ ë¬¶ì¸ ì¸ìë¥¼ ë„£ìœ¼ë©´ ìë™ìœ¼ë¡œ ì›Œë“œí´ë¼ìš°ë“œë¥¼ ì‰½ê²Œ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤

````python
def view_wordcloud(counter_class):
  # ë‹¨ì–´ ë¹ˆë„ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ WordCloudë¥¼ ìƒì„±í•¨
  wordcloud = WordCloud(background_color='white',
                      colormap = "Accent_r",width=2500, height=1500,
                      stopwords=stopwords
                      ).generate_from_frequencies(counter_class)
  plt.figure(figsize=(10,8)) # ì´ë¯¸ì§€ ì¶œë ¥ì‚¬ì´ì¦ˆ ì§€ì •
  plt.imshow(wordcloud)
  plt.tight_layout(pad=0)
  plt.axis('off')
  plt.show()

  # ìƒìœ„ 20ê°œ í‚¤ì›Œë“œì˜ ì¶œí˜„ ë¹ˆë„ìˆ˜ë¥¼ ê·¸ë˜í”„ë¡œ í‘œí˜„
  plt.figure(figsize=(10, 5)) # ì´ë¯¸ì§€ ì¶œë ¥ì‚¬ì´ì¦ˆ ì§€ì •
  plt.xlabel('keywords') # xì¶• ì´ë¦„ì§€ì •
  plt.ylabel('counts') # yì¶• ì´ë¦„ì§€ì •
  plt.xticks(rotation=70) # xì¶• í‹± ê°ë„ íšŒì „
  for key, value in zip(list(counter_class.keys())[:20],list(counter_class.values())[:20]): # ìƒìœ„ 20ê°œë§Œ ì¶”ì¶œ
      plt.bar(key, value, color='lightgrey') # barchart ìƒì„±
  plt.show()

  # ë‹¨ì–´ë³„ ë¹ˆë„ìˆ˜ ë°ì´í„° í”„ë ˆì„ ìƒì„±
  counter_df = pd.DataFrame({'word':counter_class.keys(),'counts':counter_class.values()})

  display(counter_df)
  return counter_df # return ê°’ìœ¼ë¡œ ë‹¨ì–´ë³„ ë¹ˆë„ìˆ˜ ë°ì´í„°í”„ë ˆì„ì„ ë°˜í™˜
````

````python
nouns_df = view_wordcloud(nouns_count)
````
