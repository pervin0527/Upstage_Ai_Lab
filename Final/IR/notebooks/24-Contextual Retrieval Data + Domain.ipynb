```python
import json
```

```python
# Load the two JSONL files into memory
processed_documents_queries_path = '../dataset/processed_documents_queries.jsonl'
gpt_contextual_retrieval_documents_path = '../dataset/gpt_contextual_retrieval_documents.jsonl'

# Load the files
with open(processed_documents_queries_path, 'r', encoding='utf-8') as f:
    processed_documents_queries = [json.loads(line) for line in f]

with open(gpt_contextual_retrieval_documents_path, 'r', encoding='utf-8') as f:
    gpt_contextual_retrieval_documents = [json.loads(line) for line in f]
```

```python
# Create a mapping from docid to domain from processed_documents_queries
docid_to_domain = {doc['docid']: doc['domain'] for doc in processed_documents_queries if 'docid' in doc and 'domain' in doc}

domains = set()
for k, v in docid_to_domain.items():
    print(k, v)
    domains.add(v)
```

```python
print(len(domains))

for domain in domains:
    print(domain)
```

```python
# Assign the corresponding domain to each document in gpt_contextual_retrieval_documents
for doc in gpt_contextual_retrieval_documents:
    docid = doc.get('docid')
    if docid in docid_to_domain:
        doc['domain'] = docid_to_domain[docid]

# Save the updated gpt_contextual_retrieval_documents to a new file
output_path = '../dataset/updated_gpt_contextual_retrieval_documents.jsonl'
with open(output_path, 'w', encoding='utf-8') as f:
    for doc in gpt_contextual_retrieval_documents:
        f.write(json.dumps(doc, ensure_ascii=False) + '\n')

output_path
```
